In this tutorial, we will show you how to use and setup **TriggerKit**!

### 1. Download TriggerKit

You can either install latest from the GitHub [repository](https://github.com/Datatistics/triggerkit):

```sh
$ pip install git+https://github.com/Datatistics/triggerkit.git
```

or from [pypi][pypi]


```sh
$ pip install triggerkit
```

### 2. Setup Database Connection

::: {.callout-note}
**TriggerKit** runs off of defined queries and views. Right now, we only suport connections to Snowflake, but we may add support for other databases downt the line.
:::

You can establish TriggerKit's connection to your Snowflake through your `config.toml` file:

```toml
[snowflake]
user = "your_username"
password = "your_password"  # Consider using environment variables for sensitive data
account = "your_account"
warehouse = "your_warehouse"
database = "your_database"
schema = "your_schema"
role = "ACCOUNTADMIN"  # Optional, defaults to ACCOUNTADMIN (not ideal tho...)
```
